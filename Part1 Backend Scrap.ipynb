```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import string
```

```python
response=requests.get("https://www.imdb.com/india/top-rated-indian-movies/?pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=600ca544-31f5-4bd8-ae38-ea4014c93bab&pf_rd_r=HC01NZN2WFS2HQP7SCZ6&pf_rd_s=right-4&pf_rd_t=15061&pf_rd_i=homepage&ref_=hm_india_tr_rhs_1")

html_data= response.content

list_titles=[]
list_casts=[]
list_ratings=[]

soup=BeautifulSoup(html_data,"html.parser")

run_count=0


for titles in soup.find_all("a"):
    href_tag=titles.get("href")
    if str(href_tag).startswith("/title/tt"):
        cast=titles.get("title")
        if cast==None or cast.startswith("      "):
            continue
        list_casts.append(cast)
        run_count+=1
        print("Cast: " ,cast)
        
        title=titles.get("title")
        if titles.next==" ":
            continue
        list_titles.append(titles.next)
        print("Title: ", titles.next)
        
        run_count_local=0
        
        for ratings in soup.find_all("strong"):
            rating=ratings.get("title")
            if str(rating).endswith("user ratings"):
                run_count_local+=1
                if run_count==run_count_local:
                    list_ratings.append(ratings.next)
                    print("User Ratings: ", ratings.next)
                

```

```python
df=pd.DataFrame({"Title":list_titles,"Cast":list_casts,"Rating":list_ratings})
df
```
